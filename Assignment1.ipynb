## Assignment 1: The goal of this assignment is create visuals for the Titanic dataset.

### The First section of code reads in the titanic dataset and corrects the variables Survived and Pclass to be factors.

```r
library(ggplot2)
titanic = read.csv("titanic.csv")
titanic$Survived = factor(titanic$Survived)
titanic$Pclass = factor(titanic$Pclass)
```

### From here we will create graphs in order to represent the data.

### The graph below shows the split between genders classified by who died and who survived. You can see that a higher proportion of females survived.

```r
ggplot(titanic) + geom_bar(mapping = aes(x=Sex, fill = Survived), position = "dodge")
```

### The Graph below shows the amount of people who survied and died in each Passenger class.  You can see that a passenger is more likely to survive in a higher passenger class.

```r
ggplot(titanic) + geom_bar(mapping = aes(x=Pclass, fill = Survived), position= "dodge")
```

### Scatterplot below shows the various passenger classes and how much they paid for their tickets.  Whether or not they survived is determined by their color. You can see that those in 1st class paid the highest of the 3 class groups.

```r
ggplot(data = titanic) +geom_point(mapping = aes (x=Pclass, y=Fare, color = Survived))
```

### Graph below shows percentage of passengers that paid each fare amount, split by gender. It can be observed that a higher proportion of males purchased cheap tickets. 

```r
ggplot(titanic) + geom_density(mapping = aes(x=Fare, color = Sex))
```

### BoxPlot below shows age statistics for those who survived vs those who didn't. One can see that those who died were slightly older than those who survived on average.

```r
ggplot(titanic) + geom_boxplot(mapping = aes(x = Survived, y=Age))
```

### Box Plot below shows fare distribution by gender. One can see that males purchased cheaper tickets on average.

```r
ggplot(titanic) + geom_boxplot(mapping = aes(x = Sex, y=Fare))
```

### Graph below shows the distribution of age among the titanic passengers by Gender. One can see that the highest proportion of titanic passengers were between 20 and 30.  Also a higher proportion of females were between the ages of 0 and 15 than males.

```r
ggplot(titanic) + geom_density(mapping = aes(x=Age, color = Sex))
```
